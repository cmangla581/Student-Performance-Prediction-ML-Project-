```python
# Here the ibraries are being imported 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# here we take a sample data for training and making the predictions 
data = {
    'study_hours' : [10, 5, 8, 6,12, 4, 9, 7],
    'attendance' : [95, 70, 85, 80, 98, 60, 90, 75],
    'previous_grade' : [ 88, 60, 75, 70, 92,55, 84, 68],
    'extra_activities_participation' : [1,0,1,0,1,0,1,0],
    'final_score' : [90, 65, 80, 72, 95, 58, 88, 70]
}
```

```python
# Loading the data in a table form 
df = pd.DataFrame(data)
df
```

```python
# Now comes the splitting the data into the dependent and independent features 
X = df[['study_hours', 'attendance', 'previous_grade', 'extra_activities_participation']]
y = df[['final_score']]
```

```python
X # The datset X without the final score column 
```

```python
y # the dataset of only the final score 
```

```python
# Now we train and test the dataset 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =  0.25, random_state = 42)
```

```python
X_train # The data which is being trained 
```

```python
X_test # The dataset which is being tested 
```

```python
y_train # The dataset which is being trained 
```

```python
y_test # The data which is being tested 
```

```python
# importing the Linear Regression Model 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
# Prediction of model
y_pred = model.predict(X_test) 
```

```python
# Now we evaluate the model
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Mean Absolute Error 
# Average magnitude of difference between the actual and predicted values 
print("MAE:", mean_absolute_error(y_test, y_pred)) 
```

```python
# R Squared Score 
print("R Square Score:", r2_score(y_test,y_pred))
# It indicates the proportion of a variance in a depndent variable in a regression model 
# Score nearer to 1 means a more accurate model 
```
